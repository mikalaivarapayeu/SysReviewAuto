```python
def get_data(file_name):
    sents =[]
    with open(file_name, 'r+', newline='',encoding="utf8")as f:
        reader = csv.reader(f, delimiter = '\t')
        for line in reader:
            try:
                sents.append(line)
            except Exception:
                return print(line)
        return sents
```
